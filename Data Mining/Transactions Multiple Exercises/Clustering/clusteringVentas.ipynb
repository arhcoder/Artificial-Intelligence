## Ventas por meses

```python
# Promedio de ventas por meses:
# C치lculo del promedio de ventas por mes:
dataset["Fecha"] = pd.to_datetime(dataset["Fecha"], format="%d-%m-%y")

# Suma de ventas mensuales:
ventas_mensuales = dataset.resample("M", on="Fecha")["Cantidad"].sum()

promedio_ventas_mensuales = ventas_mensuales.resample("M").mean()
sns.barplot(x=promedio_ventas_mensuales.index.strftime('%B'), y=promedio_ventas_mensuales.values)
plt.xlabel("Mes")
plt.ylabel("Promedio de Ventas")
plt.title("Promedio de Ventas por Mes")
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

## Ventas de un a침o (2014)

```python
# Filtrar los datos para un a침o espec칤fico (por ejemplo, 2014):
ventas_2014 = dataset[dataset["Fecha"].dt.year == 2014]

# Dataframe separado para cada mes:
series_mes = {}
for mes in range(1, 13):
    ventas_mes = ventas_2014[ventas_2014["Fecha"].dt.month == mes]
    series_mes[mes] = ventas_mes.resample("D", on="Fecha")["Cantidad"].sum()

# Gr치ficos de Series de Tiempo:
for mes, serie_tiempo in series_mes.items():
    plt.plot(serie_tiempo.index, serie_tiempo.values, label=f"Mes {mes}")

plt.xlabel("Fecha")
plt.ylabel("Cantidad de Ventas")
plt.title("Series de Tiempo de Ventas por Mes (2014)")
plt.xticks(rotation=90)
plt.legend(title="Mes")
plt.grid(True)
plt.show()
```

## Top de fechas en 2014

```python
import calendar

# Suma de ventas por d칤a en 2014:
ventas_diarias = ventas_2014.groupby(ventas_2014["Fecha"].dt.date)["Cantidad"].sum()

# Fechas con m치s ventas (top 10):
top10_fechas = ventas_diarias.nlargest(n=10)

# Nombre del mes para cada fecha:
nombres_meses = [calendar.month_name[date.month] for date in top10_fechas.index]
dias = [date.day for date in top10_fechas.index]

# Top 10 de fechas con ventas:
print("Fechas con m치s ventas en 2014:")
for i, fecha in enumerate(top10_fechas.index):
    print(f"{nombres_meses[i]} {dias[i]}:  {top10_fechas[i]} ventas")

# Gr치fica de top:
sns.barplot(x=top10_fechas.index, y=top10_fechas.values)
plt.xlabel("Fecha")
plt.ylabel("Cantidad de Ventas")
plt.title("Top 10 de Fechas con M치s Ventas en 2014")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## Necesidad en el negocio

**Escenario Ficticio:**

La empresa tiene como objetivo aumentar las ventas y la lealtad de los clientes. Para lograrlo, se pretenden analizar los patrones de compra y el comportamiento de los clientes y segmentarlos en grupos para desarrollar estrategias de marketing m치s efectivas.

**Objetivo del An치lisis:**

Identificar segmentos de clientes con perfiles de compra similares y desarrollar estrategias de marketing y retenci칩n personalizadas para cada segmento.

**Variables Relevantes:**
- *Cliente*: Identificaci칩n 칰nica de cada cliente.
- *Producto*: Descripci칩n de los productos comprados por cada cliente.
- *Cantidad*: Cantidad de productos comprados.
- *Precio*: Precio total de la compra.
- *Edad*: Edad de los clientes.
- *Estado_Civil*: Estado civil de los clientes.
- *Escolaridad*: Nivel de escolaridad de los clientes.
- *Sexo*: G칠nero de los clientes.
- *Dia_Semana_*: D칤a de la semana en que se realiz칩 la compra.
- *IND_Compra*: Indicador de compra.

**Ideas de An치lisis y Clusterizaci칩n:**

1. **Segmentaci칩n Demogr치fica:** Utilizando variables como *Edad*, *Estado_Civil*, *Escolaridad* y *Sexo* para segmentar a los clientes en grupos demogr치ficos.

2. **An치lisis de Comportamiento de Compra:** Utilizando variables como *Producto*, *Cantidad* y *Precio* para identificar patrones de compra.

3. **Segmentaci칩n por D칤as de la Semana:** Analizando los h치bitos de compra en funci칩n del d칤a de la semana (*Dia_Semana_*).

4. **Segmentaci칩n de Compradores Frecuentes:** Utilizando el *IND_Compra* para identificar y segmentar a los clientes que realizan compras con regularidad.

Un an치lisis interesante podr칤a ser el de segmentar a los clientes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score

# Caracter칤sticas relevantes:
selected_features = dataset[["Cantidad", "Precio", "Edad"]]

# Normalizaci칩n de los datos:
scaler = StandardScaler()
scaled_features = scaler.fit_transform(selected_features)

# Variaci칩n de los par치metros para K-means y DBSCAN:
k_values = [3, 4, 5]

# La RAM de COLAB no aguanta:
k_values = [3, 4]

best_kmeans = None
best_kmeans_silhouette = -1
best_dbscan = None
best_dbscan_silhouette = -1

# An치lisis para los diferenets par치metros:
# Ac치 se observan cu치les son la mejor combinaci칩n de par치metros con base en
# la m칠trica de silueta, con un ciclo para cada combinaci칩n de par치metros:

# PARA K-MEANS:
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans_labels = kmeans.fit_predict(scaled_features)
    silhouette = silhouette_score(scaled_features, kmeans_labels)

    if silhouette > best_kmeans_silhouette:
        best_kmeans_silhouette = silhouette
        best_kmeans = kmeans
```

```python
# Gr치fico de K-means:
plt.subplot(1, 2, 1)
plt.scatter(selected_features["Cantidad"], selected_features["Precio"], c=kmeans_labels, cmap="viridis")
plt.title(f"K-means (K={kmeans.n_clusters})")

# Gr치fico de DBSCAN:
plt.subplot(1, 2, 2)
plt.scatter(selected_features["Cantidad"], selected_features["Precio"], c=dbscan_labels, cmap="viridis")
plt.title(f"DBSCAN (eps={epsilon}, min_samples={min_samples})")

plt.tight_layout()
plt.show()

# Observaciones y diferencias:
print(f"N칰mero de cl칰steres detectados por K-means: {kmeans.n_clusters}")
print(f"N칰mero de cl칰steres detectados por DBSCAN: {len(np.unique(dbscan_labels)) - 1}")
```
