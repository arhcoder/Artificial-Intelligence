# An치lisis de Series de Tiempo

Analizando la cantidad de ventas a lo largo del tiempo en busca de patrones estacionales o tendencias.

## Ventas hist칩ricas

```python
# An치lisis general de ventas:
# Conviertiendo la fecha en tipo datetime de pandas:
dataset["Fecha"] = pd.to_datetime(dataset["Fecha"], format="%d-%m-%y")

# Suma de ventas mensuales:
ventas_mensuales = dataset.resample("M", on="Fecha")["Cantidad"].sum()

# Suma de ganancias mensuales:
ganancias_mensuales = dataset.resample("M", on="Fecha")["Importe"].sum()

# Gr치fico de ventas mensuales:
plt.figure(figsize=(12, 6))
sns.lineplot(x=ventas_mensuales.index, y=ventas_mensuales.values, label="Cantidad de Ventas")
plt.xlabel("Fecha")
plt.ylabel("Cantidad de Ventas")
plt.title("Ventas Mensuales a lo largo de los a침os")
plt.xticks(rotation=90)
plt.legend()

# Agregar la l칤nea de ganancias en el mismo gr치fico:
ax2 = plt.twinx()
sns.lineplot(x=ganancias_mensuales.index, y=ganancias_mensuales.values, color="red", label="Ganancias", ax=ax2)
ax2.set_ylabel("Ganancias")
ax2.legend(loc="upper right")

plt.grid(True)
plt.show()
```

## Descomposici칩n de series

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Descomposici칩n de la serie de tiempo de ventas mensuales:
decomposition = seasonal_decompose(ventas_mensuales, model="additive")
tendencia = decomposition.trend
estacionalidad = decomposition.seasonal
residuos = decomposition.resid

# Gr치ficos de los componentes
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(ventas_mensuales, label="Ventas Mensuales")
plt.legend(loc="upper left")
plt.subplot(412)
plt.plot(tendencia, label="Tendencia")
plt.legend(loc="upper left")
plt.subplot(413)
plt.plot(estacionalidad, label="Estacionalidad")
plt.legend(loc="upper left")
plt.subplot(414)
plt.plot(residuos, label="Residuos")
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```

```python
# Ganancias totales para cada producto:
ganancias_producto = dataset.groupby("Descripcion")["Importe"].sum()

# Producto con las ganancias m치s altas:
top_producto = ganancias_producto.idxmax()
ganancias_top_producto = ganancias_producto.max()

print(f"Producto con m치s ganancias: {top_producto}")
print(f"Ganancias totales: {ganancias_top_producto:.2f}")

# Serie de tiempo con el producto top:
datos_top_producto = dataset[dataset["Descripcion"] == top_producto]
datos_top_producto["Fecha"] = pd.to_datetime(datos_top_producto["Fecha"], format="%d-%m-%y")
ganancias_mensuales_producto = datos_top_producto.resample("M", on="Fecha")["Importe"].sum()

# Graficar las ganancias del producto a lo largo del tiempo con el nombre del producto en el t칤tulo
plt.figure(figsize=(12, 6))
sns.lineplot(x=ganancias_mensuales_producto.index, y=ganancias_mensuales_producto.values)
plt.xlabel("Fecha")
plt.ylabel("Ganancias")
plt.title(f"Ganancias del Producto {top_producto} a lo largo del tiempo")
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

## An치lisis del Queso Manchego Extramadurado NBN

```python
# Producto de inter칠s:
def product_view(product_name):
  datos_producto = dataset[dataset["Descripcion"] == product_name]
  datos_producto["Fecha"] = pd.to_datetime(datos_producto["Fecha"], format="%d-%m-%y")

  # Serie de tiempo de precios del queso:
  serie_precio = datos_producto.set_index("Fecha")["Precio"]

  # Serie de tiempo de ventas mensuales de queso:
  serie_ventas = datos_producto.resample("M", on="Fecha")["Cantidad"].sum()

  # Descomposici칩n de las Series de tiempo:
  decomposition = seasonal_decompose(serie_ventas, model="additive")
  tendencia = decomposition.trend
  estacionalidad = decomposition.seasonal

  # Graficas de series:
  plt.figure(figsize=(12, 6))
  plt.subplot(3, 1, 1)
  sns.lineplot(data=serie_precio, label="Precio")
  plt.ylabel("Precio")
  plt.title(f"Serie de Tiempo de Precio para {product_name}")

  plt.subplot(3, 1, 2)
  sns.lineplot(data=serie_ventas, label="Ventas Mensuales")
  plt.ylabel("Cantidad de Ventas")
  plt.title(f"Serie de Tiempo de Ventas para {product_name}")

  plt.subplot(3, 1, 3)
  sns.lineplot(data=tendencia, label="Tendencia")
  sns.lineplot(data=estacionalidad, label="Estacionalidad")
  plt.ylabel("Valor")
  plt.title(f"Descomposici칩n de Ventas Mensuales para {product_name}")
  plt.legend(loc="upper left")
  plt.tight_layout()
  plt.show()

product_view(top_producto)
```

## Crecimiento de precios

```python
# Crecimiento de precios en la tienda:
dataset["Fecha"] = pd.to_datetime(dataset["Fecha"], format="%d-%m-%y")
dataset.sort_values(by="Fecha", inplace=True)

# Agrupamiento por promedio de precio mensual:
promedio_mensual = dataset.resample("M", on="Fecha")["Precio"].mean()
crecimiento = promedio_mensual.pct_change()

# Serie de tiempo:
sns.lineplot(x=crecimiento.index, y=crecimiento.values)
plt.xlabel("Fecha")
plt.ylabel("Crecimiento Precio Promedio Mensual")
plt.title("Crecimiento del Precio Promedio Mensual en el Dataset")
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
