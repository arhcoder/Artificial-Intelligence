# Regresión sobre una función polinomial

El objetivo es seleccionar la función base y usar mínimos cuadrados para aprender el peso apropiado para encontrar la línea que mejor se ajuste a los datos.

```python
import numpy as np
import scipy.sparse as sp
from math import sqrt
import matplotlib.pyplot as plt
```

## Conjunto de datos


```python
data = np.loadtxt("ML06.txt")
X_b = data[:,0]
Y_b = data[:,1]

plt.scatter(X_b, Y_b)
plt.show()
```

```python
X_N = np.sin(X_b).reshape(X_b.shape[0],1)
ones_column = np.ones((X_N.shape[0], 1))
X_ = np.hstack((ones_column, X_N))

theta = np.linalg.inv(X_.T @ X_) @ (X_.T @ Y_b)

x_values = np.linspace(-10, 10, 100)
y_values = theta[1] * np.sin(x_values) + theta[0]

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, color="red")
plt.scatter(X_b, Y_b, color="blue")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
sse = lambda yt, yp: ((yt - yp)**2).mean()
```

```python
ypred = theta[1] * np.sin(X_b) + theta[0]
```

```python
sse = sse(Y_b, ypred)
print("Error:", sse)
```
