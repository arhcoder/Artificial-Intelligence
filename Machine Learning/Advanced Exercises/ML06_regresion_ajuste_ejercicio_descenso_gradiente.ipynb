# Regresión sobre una función polinomial

El objetivo es seleccionar la función base y usar mínimos cuadrados para aprender el peso apropiado para encontrar la línea que mejor se ajuste a los datos.

```python
import numpy as np
import scipy.sparse as sp
from math import sqrt
import matplotlib.pyplot as plt
```

## Conjunto de datos


```python
data = np.loadtxt("ML06.txt")
X_b = data[:,0]
Y_b = data[:,1]

plt.scatter(X_b, Y_b)
plt.show()
```

```python
# Descenso del gradiente
theta = np.array([1.0])
learning_rate = 0.001
num_iterations = 10000
for i in range(num_iterations):
    Y_pred = theta[0] * np.sin(X_b)
    error = Y_pred - Y_b
    gradient = 2 * np.dot(error, np.sin(X_b))
    theta -= learning_rate * gradient

# Grafica:
X_plot = np.linspace(min(X_b), max(X_b), 100)
Y_plot = theta[0] * np.sin(X_plot)
plt.scatter(X_b, Y_b, color="blue")
plt.plot(X_plot, Y_plot, color="red")
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
