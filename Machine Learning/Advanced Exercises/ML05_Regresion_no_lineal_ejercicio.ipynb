# Mínimos cudrados para una función cuadrática

Existen problemas en los que no necesariamente el comportamiento de los variables es necesariamente lineal. En este ejercicio se busca mostrar el uso de mínimos cuadrados ordinarios para una función cuadrática.

**Objetivo:** Encontrar la línea que mejor ajuste a los datos empleando la solución de mínimos cuadrados.

Suponemos un polinomio de grado dos con un atributo, esto es,

$$
y_{i}=f_{\mathbf{\theta}}(x) = \theta_{1}x_{1}^{2}+\epsilon,
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# generamos datos artificialmente
X = np.random.normal(0, 1, size=(100,1))
Y = 2 * np.power(X,2) + np.random.normal(0,1,size=(100,1))

plt.scatter(X, Y, color="blue")
plt.show()
```

```python
# Ajustando con una función cuadrática:
X_sq = np.power(X, 2)
X_b = np.column_stack((X_sq, np.ones_like(X_sq)))

theta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ Y

# Generamos puntos para graficar la función ajustada
X_plot = np.linspace(-3, 3, 100)
Y_plot = theta[0][0] * np.power(X_plot, 2)

# Graficamos los datos y la función ajustada
plt.scatter(X, Y, color="blue", label="Datos")
plt.plot(X_plot, Y_plot, color="red")
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
