```python
import pandas as pd
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
```

```python
# Obtiene un dataset de ejemplo (los originales sin preprocesamiento):
dataset = pd.read_csv("/content/drive/MyDrive/IIMAS/ML-DATASET/EMOTIV/profiles/Alex1/dataset.csv")
serieArbol = dataset.loc[dataset["Expected Output"] == "Arbol"]
seconds = np.arange(len(serieArbol)) / 250

plt.figure(figsize=(20, 10))
plt.plot(seconds, serieArbol["Channel 3"])
plt.xticks(ha="right")
plt.title("Time Serie on EEG Channel 3 from \"profiles/Alex1\"")
plt.xlabel("Segundos")
plt.ylabel("Hz Intensity")
plt.show()
```

```python
# Obtiene un dataset de ejemplo (normalizado):
dataset = pd.read_csv("/content/drive/MyDrive/IIMAS/ML-DATASET/EMOTIV/normalizated/Alex1/dataset.csv")
serieArbol = dataset.loc[dataset["Expected Output"] == "Arbol"]
seconds = np.arange(len(serieArbol)) / 250

# Grafica los 14 canales:
plt.figure(figsize=(20, 10))
plt.xticks(ha="right")
plt.title("Time Serie on EEG Channels from \"normalizated/Alex1\"")
plt.xlabel("Segundos")
plt.ylabel("Normalizated Hz")
for channel in range(13):
  plt.plot(seconds, serieArbol[f"Channel {channel+1}"])
plt.show()
```
