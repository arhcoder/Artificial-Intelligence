# [04] üéé Apriori

## Patrones de Pasajeros

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Haciendo bins para la edad:
dataset["Age_Group"] = pd.cut(dataset["Age"], bins=[0, 18, 36, 60, 100], labels=["child", "adult", "middle", "senior"])

# Obteniendo duumies con Pclass, Sex y Age_Group:
encoded_data = pd.get_dummies(dataset[["Pclass", "Sex", "Age_Group"]], columns=["Pclass", "Sex", "Age_Group"])

# Convirtiendo los datos en transacciones:
transactions = encoded_data.values

# Algoritmo Apriori:
min_support = 0.1
frequent_itemsets = apriori(pd.DataFrame(transactions, columns=encoded_data.columns), min_support=min_support, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
print(rules[["antecedents", "consequents", "support"]])
```

## Asociaciones en Sobrevivientes

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Nuevos dummies incluyendo el "Survived":
encoded_data = pd.get_dummies(dataset[["Pclass", "Sex", "Age_Group", "Survived"]], columns=["Pclass", "Sex", "Age_Group", "Survived"])

# Convirtiendo en transacciones:
transactions = encoded_data.values

# Algoritmo Apriori:
min_support = 0.1
frequent_itemsets = apriori(pd.DataFrame(transactions, columns=encoded_data.columns), min_support=min_support, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

# Se filtra a s√≥lo las asociaciones que sean sobre sobrevivir o no:
filtered_rules = rules[
    (rules["consequents"].apply(lambda x: any(item in str(x) for item in ["Survived_0", "Survived_1"])))
]
ordered_rules = filtered_rules.sort_values(by="support", ascending=False)
print(ordered_rules[["antecedents", "consequents", "support"]])
```
